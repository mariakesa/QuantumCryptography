```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
p=np.arange(0,100)
x=np.arange(0,100)

def Hamiltonian(p,x,l=10,k=0.5):
    h_arr=np.zeros((p.shape[0],x.shape[0]))
    for p_ in range(p.shape[0]):
        for x_ in range(x.shape[0]):
            h_arr[p_,x_]=p_**2/2+k*(x_-l)**2/2
    return h_arr
X, Y = np.meshgrid(p, x)
h=Hamiltonian(p,x)

fig = plt.figure()
ax = fig.gca(projection='3d')
# Plot the surface.
surf = ax.plot_surface(X, Y, h.T, cmap=cm.coolwarm,
                       linewidth=0, antialiased=True)
ax.set_xlabel('Momentum p')
ax.set_ylabel('Position x')
```
